```python
### Basic color graph
from queue import Queue

def color_graph(graph):
    n = len(graph)
    color = [0 for i in range(n)]
    #result = True
    for i in range(n):
        if color[i] == 0:
            Q = Queue()
            Q.put(i)         # add ?
            color[i] = 1
            while not Q.empty():
                v = Q.get()  # pop ? 
                for u in graph[v]:
                    if color[v] == color [u]:
                        return False
                    if color[u] == 0:
                        color[u] = color[v]%2 + 1
                    Q.put(u)
    return color

graph = [[1, 2], [0, 2], [0, 1]]
color_graph(graph) 
```

```python

#Вопрос 2 Correct
#Given an adjacency list for a graph determine if a graph is bipartite. 
#(A graph is bipartite if it's vertices can be divided in to two 
# sets such that there is no edges between vertices that are 
#in the different sets).


from queue import Queue
from collections import deque

def isBipartite(adj_list):
    n = len(adj_list)
    is_bipartite: bool = True
  
    colour = [0 for i in range(n)]
      
    for i in range(n):
        if colour[i] == 0:
            Q = deque()
            Q.append(i)         
            colour[i] = 1
            
            while Q:
                v = Q.popleft()  
                for u in adj_list[v]:
                    if colour[v] == colour [u]:
                        is_bipartite = False
                    if colour[u] == 0:
                        colour[u] = colour[v] % 2 + 1
                        Q.append(u)
    return is_bipartite



adj_list = [[], [2], [1]]
isBipartite(adj_list)
```

```python
### Task 3. Correct

#You're given a maze map where '.' are empty cells, 
#'#' are impassable wall cells, 'X' is the exit. 
#Construct its adjacency list. Vertices are non-wall cells, 
#edges connect cells that share a side. 

def maze2graph(maze):
    height, width = len(maze), len(maze[0])
    graph = {(x, y): [] for y in range(width) for x in range(height) if maze[x][y] != "#"}

    for row, col in graph.keys():
        if row < height - 1 and maze[row + 1][col] != '#':
            graph[(row, col)].append((row + 1, col))
            graph[(row + 1, col)].append((row, col))
        if col < width - 1 and maze[row][col + 1] != '#':
            graph[(row, col)].append((row, col + 1))
            graph[(row, col + 1)].append((row, col))

    return graph

maze = ['....#.']
maze2graph(maze)
```
