HERO vs MONSTER (separated functions)

In this game we have 2 objects: HERO and MONSTER. A hero has to go throught a field (400-cell field) to collect hit power, different arms, lifes, abilities and so on. Some where at the end of this way, he/she has to meet and to fight with a monster. They will beat each over and the winner will remain alife.

Functions game_field_hero() and game_field_monster() create 2 matrices. The first (field_h) contains heros location, second (field_m) containes monster and collections location. Heros is 1, monster is 3 and findings (we can change the hero's abilities) are 2. Both, field_h and field_m are global variables.

```python
import numpy as np
import random

def game_field_hero():
    global r
    r = 0
    global c 
    c = 0    
    field = np.zeros((20, 20)) # To create alternativ matrix 20 per 20 with 0's for hero
    field[r][c] = 1 # heros location
    global field_h
    field_h = field
    return field_h

def game_field_monster():
    global field_m
    field = np.zeros((20, 20)) # To create a matrix wih 0's 20 per 20 with 0's for collection and monster
    field [random.randint(15, 18)][random.randint(15, 18)] = 3 # monsters location  
    for i in range (10):
        field [random.randint(0, 15)][random.randint(0, 15)] = 2 # collect hit power, different arms, lifes, abilities...
    global field_m
    field_m = field
    return field_m  

print(game_field_hero())
print(game_field_monster())
```

Function hm_caracteristics(a) generates the initial hero and monster abilities depends on the selected level a (from 1 to 4)


The MONSTER's caracteristics : LIFES - 200 
                              ATTACKS - (hand, foot, sword_1, sword_2, magic attacks type) 
                              DEFEND - % reduction from hero hit
                              
The variable monster_car = [LIFES, ATTACKS[hand, foot, sword, super attacks type], DEFEND]   


The HERO's caracteristics: LIFES = 250 can be completed max 270 
                           ATTACKS (hand, foot, sword_1, sword_2, magic atacks type) 
                           DEFEND  % reduction from monster hit 
                           INSPIRATION (inspiration 1, 2, 3, 4, 5 means +1, +2, +3, +4, +5 for hero's hit power) 
                           DEXTERITY (dexterity 1, 2, 3, 4, 5 means -1, -2, -3, -4, -5 from monster's hit power)
                           
The variable hero_car = [LIFES, ATTACKS[hand, foot, sword_1, sword_2, magic atacks type], DEFEND, INSPIRATION, DEXTERITY] 

LIFES, INSPIRATION and DEXTERITY can be increaesd during the navigation

Both variables are global. 


```python
def hm_caracteristics (a): # depends on the level a to determine the initial hero's and monsters caracteristics
    global hero_car
    hero_car = [250, [2, 3, 5, 10, 15], 0.2, 1, 1]
    global monster_car
    if a == 1: monster_car = [200, [4, 6, 8, 10, 20], 0.1]
    elif a == 2: monster_car = [200, [4, 6, 8, 10, 20], 0.2]
    elif a == 3: monster_car = [200, [4, 6, 8, 10, 20], 0.3]   
    elif a == 4: monster_car = [200, [4, 6, 8, 10, 20], 0.4]
    return monster_car, hero_car

a = int(input("Enter, please, a level from 1 to 4: "))

hm_caracteristics (a)

hero_car
```

Function upgrade() is destinated to change the heros abilities. There are 10 cell with value "2". In condition if heros location in field_h has the same coordinates as "2" in field_h we randomly generate a number from 1 to 4 and depends on this number a catacteristic changes.

```python
def upgrade(): # to modify the heros characteristics
    global hero_car
    global r
    global c
    if int(field_m[r, c]) == 2:
        b = random.randint(1, 4)
        print(b)
        if b == 1 and (hero_car[0]+20) <= 270: 
            hero_car[0] = hero_car[0] + 20
            return hero_car
        elif b == 2: 
            hero_car[2] = hero_car[2] + 0.2
            return hero_car
        elif b == 3 and (hero_car[3]+1) <= 5: 
            hero_car[3] = hero_car[3] + 1
            return hero_car
        elif b == 4 and (hero_car[4]+1) <= 5: 
            hero_car[4] = hero_car[4] + 1
            return hero_car

print(upgrade())
print("The heros caracteristics after updating", hero_car) # the hero's caracteristic after each step 
```

Function fighting() contains hero's and monster's hits one by one until the result of hero's or monster's wins. The hits formulas take in account the participant caracteristics

```python
def fighting():
    while  hero_car[0] >= 1 and monster_car[0] >=1:
        # hero's hit
        c = random.randint(0, 4)
        monster_car[0] = monster_car[0] - (int(hero_car[1][c]) + hero_car[3])*(1 - monster_car[2])
        print("monster_car lifes = ", monster_car[0])
        # monster's hit
        d = random.randint(0, 4)
        hero_car[0] = hero_car[0] - (int(monster_car[1][d]) - hero_car[4])*(1 - hero_car[2])
        print("hero_car lifes = ", hero_car[0])
    if hero_car[0] < 1: return "The Monster Wins"
    else: return "The Hero Wins"

if (int(field_m[r, c]) == 3):
    print(fighting())
```
