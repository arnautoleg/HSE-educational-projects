```python
import numpy as np
from copy import deepcopy
from random import seed, randint
import time
import math


def find_percentile(a, b, p):
    total_lenght = len(a) + len(b)
    half = total_lenght * p // 100
    
    
    if len(b) < len(a):
        a, b = b, a
    
    l, r = 0, len(a) - 1
    while True:
        i = (l + r) // 2
        j = half - i - 2
        Aleft = a[i] if i >= 0 else float("-infinity")
        Aright = a[i + 1] if (i + 1) < len(a) else float("infinity")
        Bleft = b[j] if j >= 0 else float("-infinity")
        Bright = b[j + 1] if (j + 1) < len(b) else float("infinity")

        if Aleft <= Bright and Bleft <= Aright:
            if total_lenght % 2:
                return min(Aright, Bright)
            return min(max(Aleft, Bleft), min(Aright, Bright))
        elif Aleft > Bright:
            r = i - 1
        else:
            l = i + 1       
            
def alternative(list_1, list_2, prob):
    a = deepcopy(list_1)
    b = deepcopy(list_2)
    p = deepcopy(prob)

    result = a + b
    #print("result", result)
    result.sort()
    #print (int(math.ceil(len(result)*p/100)) - 1)
    return result[int(math.ceil(len(result)*p/100)) - 1]
            

def test_find_percentile(a, b, p, answer):
    #sorted_test = mergesort(test)
    sorted_test = find_percentile(a, b, p)
    error_str = "Test failed!\nInput: {0}\nOutput: {1}\n Correct output: {2}"
    assert sorted_test == answer, error_str.format([a, b, p], sorted_test, answer)
    #print("Unit test passed !")
            
def run_unit_test():
    test_find_percentile([1, 2, 7, 8, 10], [5, 6, 12], 50, 6)
    test_find_percentile([1, 2, 7, 8, 10], [6, 12], 50, 7)
    test_find_percentile([1, 2, 7, 8], [6, 12], 50, 6)
    test_find_percentile([15, 20, 35, 40, 50], [], 50, 35)
    test_find_percentile([15, 20], [25, 40, 50], 50, 25)
    print("Unit test passed !")
    
def run_stress_test(max_test_size=10, max_attempts=5, max_right_border=20):
    seed(100)
    for test_size in range(1, max_test_size):
        print('test_size = ', test_size)
        for right_border in range(1, max_right_border):
            for attempt in range(max_attempts):
                p = 50
                a = list(np.random.randint(0, right_border, test_size))
                b = list(np.random.randint(0, right_border, test_size))
                a.sort()
                b.sort() 
                test_find_percentile(a, b, p, alternative(a, b, p))
    print('Stress test passed!')

def run_max_test():
    percentile = 50
    random_size = np.random.randint(1, 150000)
    a = list(np.random.randint(1, 10000, random_size))
    random_size = np.random.randint(1, 150000)
    b = list(np.random.randint(1, 10000, random_size))
    a.sort()
    b.sort()
    
    start = time.time()
    result = find_percentile(a, b, percentile)
    end = time.time()
    print("my_solution_time =", (end - start))
     
    
    if result == alternative(a, b, percentile):
        print('Stress test passed!')
    else:
        print('Stress test failed!')  
    
def main():
    run_unit_test()
    run_stress_test()
    run_max_test()

if __name__ == "__main__":
    main()      
```
