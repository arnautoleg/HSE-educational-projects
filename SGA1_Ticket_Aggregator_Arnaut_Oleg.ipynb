```python
### Data introduction

# Program asks an positive number for price 
# Controls if this is a number, if not user has to enter the data again 

ticket_price = input("Enter the price (positive number): ")

try: a = float(ticket_price)
except: a = "not number"
    
while a == "not number" or a <= 0:
    print("Please, enter the price one more time")
    ticket_price = input("Enter the price (positive number): ")
    try: a = float(ticket_price)
    except: a = "not number" 
ticket_price = float(ticket_price)

# Program asks an positive number for transfers 
# Controls if this is a number, if not user has to enter the data again     

transfers_number = input("Enter the transfer number (positive integer number): ") 

try: b = int(transfers_number)
except: b = "not number"
    
while b == "not number" or b < 0 or (float(transfers_number) != int(float(transfers_number))):
    print("Please, enter the transfer number one more time")
    transfers_number = input("Enter the transfer number (positive integer number): ")
    try: 
        b = int(transfers_number)
        transfers_number = b
    except: b = "not number" 
transfers_number = int(transfers_number)

# Program asks an positive number for duration 
# Controls if this is a number, if not user has to enter the data again  

hours = input("Enter the duration in hours(positive number): ")  

try: c = float(hours)
except: c = "not number"
    
while c == "not number" or c <= 0:
    print("Please, enter the duration one more time")
    hours = input("Enter the duration in hours(positive number): ")
    try: c = float(hours)
    except: c = "not number"      
hours = float(hours)

# Program asks str input with instructions for refund 
# Controls if this correct entered, if not user has to enter the data again 

refund = input(""""The refund, type "included" or "not included": """)  

while refund != "included" and refund != "not included":
    print("Please, enter the refund one more time")
    refund = input(""""The refund, please type "included" or "not included": """)
    
# Program asks str input with instructions for luggage
# Controls if this correct entered, if not user has to enter the data again    

luggage = input(""""The luggage, type "included" or "not included": """) 

while luggage != "included" and luggage != "not included":
    print("Please, enter the luggage one more time")
    luggage = input(""""The luggage, please type "included" or "not included": """)
    
# Program asks str input with instructions for refund need
# Controls if this correct entered, if not user has to enter the data again   

need_refund = input("""Do you need refund? Type "Yes" or "No" : """) 

while need_refund != "Yes" and need_refund != "No":
    print("Please, enter the need_refund one more time")
    need_refund = input("""Do you need refund? Type "Yes" or "No" : """)

# Program asks str input with instructions for luggage need
# Controls if this correct entered, if not user has to enter the data again 

need_luggage = input("""Do you need luggage? Type "Yes" or "No" : """)   

while need_luggage != "Yes" and need_luggage != "No":
    print("Please, enter the need_luggage one more time")
    need_luggage = input("""Do you need luggage? Type "Yes" or "No" : """) 
  
    
### Classification block  

# Each parameter is estimateted by categories and each category has a score   
    
if float(ticket_price) < 200: ticket_value = 1   # the ticket price is the most important criteria 
elif 200 <= float(ticket_price) < 250: ticket_value = 4                    
elif 250 <= float(ticket_price) < 300: ticket_value = 7  
else: ticket_value = 10    
                    
if hours < 10: hours_value = 1            # the duration is the second criteria
elif 10 <= hours < 20: hours_value = 3                    
elif 20 <= hours < 30: hours_value = 5  
else: hours_value = 7   
                    
if transfers_number == 0: transfers_number_value = 1 # as usual is related to duration, but it is not obligatory
elif 1 <= transfers_number <= 2: transfers_number_value = 2                    
elif transfers_number == 3: transfers_number_value = 3  
else: transfers_number_value = 4

# For flexibility increasing the values of luggage and refund depends on users needs

if need_luggage == "No" or (need_luggage == "Yes" and luggage == "included"): need_luggage_value = 0 
else: need_luggage_value = 1           

if need_refund == "No" or (need_refund == "Yes" and refund == "included"): need_refund_value = 0
else: need_refund_value = 1            

# Intergral score estimation, we can try to take all parameters toghether each parameter having his own weight

score = ticket_value + hours_value + transfers_number_value + need_luggage_value + need_refund_value 

# Offer category determination and Offers caracteristics printing

if 3 <= score <= 6: print('Offer category:', 'very good')
elif 7 <= score <= 12: print('Offer category:', 'good')
elif 13 <= score <= 18: print('Offer category:', 'bad')
elif score > 18: print('Offer category:', 'very bad')                           
                    
print (f"Offer: Ticket price - {ticket_price} $, Duration - {hours} hours,\
 Transfers number - {transfers_number}, Refund - {refund }, Luggage - {luggage}")
```
