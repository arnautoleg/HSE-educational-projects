```python
from heapq import heappop, heappush  

# heappop - remove top element from the heap
# heappush - ading element to the heap

def dijkstra(adj_matrix, v_from, v_to):
    n, graph = len(adj_matrix), adj_matrix
    dist = [float("inf") for i in range(n)] # n - number of vertices in graph
    heap = []
    visited = [False for i in range(n)]
    
    heappush(heap, (0, v_from))  # distance and the first element we put in heap
    dist[v_from] = 0
    
    graph_dict = {}
    
    #for i in range(0, n):
        #graph_dict[i] = [n*[]]
        #for j in range(0, n):
            #print(graph[i][j])
            #graph_dict[i][j] = graph[i][j]
    
    graph_dict = []
    for i in range (0, n):
        for j in range (0, n):
            graph_dict.append([[i, j], graph[i][j]])
   
    #print(graph_dict)
    #print(graph_dict[1][4])
    
    while len(heap) > 0:
        d, v = heappop(heap) # d - distance and v - current vertex that is not visited yet
        #print("d =", d)
        
        visited[v] = True
        
        print(visited)
        if dist[v] < d:
            continue
        #print(graph_dict[0][0][0])
        #print(v)
        #print(graph[v][1])
        for u, c in graph_dict:  # U - pair of vertices, c - costs (adj list)
            #print(u, c)
            #print("dist[u[1]] = ", dist[u[1]])
            if not visited[u[1]] and d + c < dist[u[1]] and graph_dict[0][0][0] == v:
                print(d, u, c)
                dist[u[1]] = d + c
                heappush(heap, (dist[u[1]], u[1]))
                  
    return dist

adj_matrix = [[float('inf'), 4, 10, float('inf'), float('inf')],
              [float('inf'), float('inf'), 1, 1, 7], 
              [float('inf'), float('inf'), float('inf'), float('inf'), 2],
              [float('inf'), float('inf'), float('inf'), float('inf'), 1],
              [float('inf'), 7, float('inf'), 8, float('inf')]
             ]

v_from, v_to = 0, 2
dijkstra(adj_matrix, v_from, v_to)
```
